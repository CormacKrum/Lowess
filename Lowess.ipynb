```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
y = y.reshape((-1,1))
X = (X-X.mean())/X.std()
```

```python
plt.figure()
plt.scatter(X,y)
plt.title("Normalized Data")
plt.show()
```

```python
loss = (X-y)**2
```

```python
plt.plot(loss)
plt.title("Loss")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((ones,X))
```

```python
theta = np.dot(np.dot(np.linalg.inv(np.dot(X_.T,X_)),X_.T),y)
```

```python
def predict(X,theta):
    return np.dot(X,theta)
```

```python
y_pred = predict(X_,theta)
```

```python
def R2_Score(y,y_pred):
    num = np.sum((y-y_pred)**2)
    den = np.sum((y-y.mean())**2)
    return (1-num/den)*100
```

```python
R2_Score(y,y_pred)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(X,y_pred,c="red",label="prediction line")
plt.legend()
plt.show()
```
